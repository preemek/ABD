```python
import pandas as pd
#https://www.jeannicholashould.com/tidy-data-in-python.html
```

```python
#ORIGINAL DATA
df = pd.read_csv("earthquake_data.csv")
```

```python
df
```

```python
#ANALYSIS DATA
formatted_df = pd.melt(df, id_vars=['Age', 'What is your gender?'], value_vars=['Do you think the "Big One" will occur in your lifetime?'], value_name='Answer', var_name='Question for this person')

formatted_df = formatted_df.dropna()

formatted_df = formatted_df.sort_values(ascending=True, by=["Age"])

formatted_df
```
